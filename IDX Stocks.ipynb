```python
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
sns.set_style('whitegrid')

import matplotlib.pyplot as plt
import warnings
import plotly
import cufflinks as cf
%matplotlib inline
cf.go_offline()
```

# Data 
We will get stock information for the following banks:
- BCA
- Mandiri
- BNI
- BRI

```python
BBCA = pd.read_excel('Finance Project IDX.xlsx',header=0, sheet_name='BBCA')
BBCA['Date'] = BBCA['Date'].dt.date #convert date time to date 
BBCA.set_index('Date',inplace=True)
BBCA
```

```python
BBNI = pd.read_excel('Finance Project IDX.xlsx',header=0, sheet_name='BBNI')
BBNI['Date'] = BBNI['Date'].dt.date #convert date time to date 
BBNI.set_index('Date',inplace=True)
BBNI
```

```python
BMRI = pd.read_excel('Finance Project IDX.xlsx',header=0, sheet_name='BMRI')
BMRI['Date'] = BMRI['Date'].dt.date #convert date time to date 
BMRI.set_index('Date',inplace=True)
BMRI
```

```python
BBRI = pd.read_excel('Finance Project IDX.xlsx',header=0, sheet_name='BBRI')
BBRI['Date'] = BBRI['Date'].dt.date #convert date time to date 
BBRI.set_index('Date',inplace=True)
BBRI
```

```python
tickers = ['BBCA','BMRI','BBNI','BBRI']
```

```python
bank_stocksid = pd.concat([BBCA,BMRI,BBNI,BBRI],keys=tickers,axis=1)
```

```python
bank_stocksid.columns.names = ['Bank','Stock Info']
```

# Returns

```python
returns = pd.DataFrame()
```

```python
for bank in tickers:
    returns[bank + "return"]=bank_stocksid[bank]["Close"].pct_change()
```

```python
returns.head()
```

```python
sns.set_style('whitegrid')
sns.pairplot(returns.dropna())
```

**The dates which each bank stock had the best single day returns**

```python
returns.idxmax()
```

**The dates which each bank stock had the worst single day returns**

```python
returns.idxmin()
```

**Standard Deviation**

```python
returns.std()
```

**Riskiest stock for the year 2020**

```python
startdate = datetime.datetime.strptime('2020-01-01', "%Y-%m-%d").date()
enddate = datetime.datetime.strptime('2020-12-31', "%Y-%m-%d").date()
```

```python
returns.loc[startdate:enddate].std()
```

# 2020 returns

```python
f, axes = plt.subplots(2,2,figsize=(15,11),sharex=True)
sns.despine(left=True)

#Plot BBCA 
sns.distplot(returns.loc[startdate:enddate]['BBCAreturn'].dropna(),bins=80,ax=axes[0, 0])

#Plot BMRI
sns.distplot(returns.loc[startdate:enddate]['BMRIreturn'].dropna(),bins=80,ax=axes[0, 1])

#Plot BBNI
sns.distplot(returns.loc[startdate:enddate]['BMRIreturn'].dropna(),bins=80,ax=axes[1, 0])

#Plot BBRI
sns.distplot(returns.loc[startdate:enddate]['BBRIreturn'].dropna(),bins=80,ax=axes[1, 1])

```

# Moving Average 
moving average is used to see the stock trend

```python
plt.figure(figsize=(12,4))
BBCA['Close'].loc[startdate:enddate].rolling(window=30).mean().plot(label="30 Day Avg")
BBCA['Close'].loc[startdate:enddate].plot(label="BCA Close")
plt.legend()
```
